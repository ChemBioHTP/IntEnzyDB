## This function I wrote is to extract only the seqres as a list

```python
def get_seqres(pdb_id):
    """ Return the seqres sequence of a pdb file
    
    >>> get_seqres('4Z0L')
    >>> get_seqres('4lza')
    """
    pdb_file = get_pdb_file(pdb_id, filetype='pdb', compression=False) 
    # using a get_pdb_file() function from pypdb package to return a file with format 'pdb'.
    file1 = pdb_file.splitlines()
    # split this long string into list by \n.
    list_se = []
    for line in file1:
        if line[:6] == "SEQRES":
            list_se.append(line)
    return(list_se)
```

## We first used pandas to convert a list into dataframe 

### First we used split() to split each string in the list returned by get_atom() function 

```python
import pandas as pd
```

```python
def get_atom(pdb_id):
    """ Return the atom sequence of a pdb file as a pandas dataframe
    
    >>> get_atom('4Z0L')
    >>> get_atom('4lza')
    """
    pdb_file = get_pdb_file(pdb_id, filetype='pdb', compression=False)
    # using a get_pdb_file() function from pypdb package to return a file with format 'pdb'.
    file1 = pdb_file.splitlines()
    list_atom = []
    for line in file1:
        if line[:4] == "ATOM":
            list_atom.append(line)
    list_s_atom = [s.split() for s in list_atom]
    # split each string in a list by white spaces
    df = pd.DataFrame(list_s_atom)
    # use DataFrame function to convert a list to dataframe
    df["id"] = pdb_id
    # add one id column to exsiting dataframe
    return(df)
```

```python
def get_all_atom(pdb_id):
    """ Return the all atom sequence of a pdb file as a pandas dataframe
    
    >>> get_all_atom('4Z0L')
    >>> get_all_atom('4lza')
    """
    pdb_file = get_pdb_file(pdb_id, filetype='pdb', compression=False)
    # using a get_pdb_file() function from pypdb package to return a file with format 'pdb'.
    file1 = pdb_file.splitlines()
    list_atom = []
    
    # Find the index number of the line with ATOM
    for line in file1:
        if line[:4] == "ATOM":
            num = file1.index(line)
            break
    
    # we add/append all the lines which are within atom part until we meet CONECT
    while file1[num][:6] != "CONECT":
        list_atom.append(file1[num])
        num = num + 1
        
    list_s_atom = [s.split() for s in list_atom]
    # split each string in a list by white spaces
    
    df = pd.DataFrame(list_s_atom)
    # use DataFrame function to convert a list to dataframe
    df["id"] = pdb_id
    # add one id column to exsiting dataframe
    return(df)
```

```python
import warnings
```

```python
class pdb:
    """A class of each pdb_id file, including desired attributes and methods"""
    
    def __init__(self, pdb_id):
        """Create a object given a specific id"""
        
        self.id = pdb_id
        self.file = get_pdb_file(pdb_id, filetype='pdb', compression=False)
        if self.file == None:
            warnings.warn("This pdb id file doesn't have a pdb format file")
            self.list = None
        else:
            self.list = self.file.splitlines()
        
    def get_enzyme_type(self):
        """get the enzyme type of this pdb"""
        
        if self.list == None:
            warnings.warn("This pdb id file doesn't have a pdb format file")
            return None
        for line in self.list:
            if line[:6] == "HEADER":
                lis = line.split()[1:]
                res = ""
                for i in lis:
                    if not re.findall("^[0-3][1-9]-[A-Z]{3,}-[0-9][0-9]$", i):
                        res = res + i + " "
                    else:
                        break
                return(res.rstrip())
        warnings.warn("This pdb file doesn't have enzyme type.")
            
    def get_name(self):
        """get molecue name of this pdb"""
        
        if self.list == None:
            warnings.warn("This pdb id file doesn't have a pdb format file")
            return None
        for line in self.list:
            if "MOLECULE:" in line:
                res = line[line.find("MOLECULE:")+10:]
                if ";" in res:
                    return(res[:res.find(";")])
                else:
                    res = res.rstrip() + self.list[self.list.index(line)+1][11:]
                    return(res[:res.find(";")])
        warnings.warn("This pdb file doesn't have name field.")
            
    def get_organism_name(self):
        """get organism name of this pdb"""
        
        if self.list == None:
            warnings.warn("This pdb id file doesn't have a pdb format file")
            return None
        for line in self.list:
            if "ORGANISM_SCIENTIFIC:" in line:
                res = line[line.find("ORGANISM_SCIENTIFIC:")+21:]
                return(res[:res.find(";")])
        warnings.warn("This pdb file doesn't have organism name field.")
            
    def get_organism_taxid(self):
        """get organism taxid of this pdb"""
        
        if self.list == None:
            warnings.warn("This pdb id file doesn't have a pdb format file")
            return None
        for line in self.list:
            if "ORGANISM_TAXID:" in line:
                res = line[line.find("ORGANISM_TAXID:")+16:]
                return(res[:res.find(";")])
        warnings.warn("This pdb file doesn't have organism tax id field.")
            
    def get_chain_id(self):
        """get chain id of this pdb"""
        
        if self.list == None:
            warnings.warn("This pdb id file doesn't have a pdb format file")
            return None
        chain_id = ""
        for line in self.list:
            if "CHAIN:" in line:
                res = line[line.find("CHAIN:")+7:]
                chain_id += res[:res.find(";")] + ", "
        if chain_id:
            return chain_id.rstrip(", ")
        warnings.warn("This pdb file doesn't have chain id field.")
            
    
    def get_ec_number(self):
        """get EC number of this pdb"""
        
        if self.list == None:
            warnings.warn("This pdb id file doesn't have a pdb format file")
            return None
        for line in self.list:
            if "EC:" in line:
                res = line[line.find("EC:")+4:]
                return(res[:res.find(";")])
        warnings.warn("This pdb file doesn't have EC number field.")
            
    def get_strain(self):
        """get strain of this pdb"""
        
        if self.list == None:
            warnings.warn("This pdb id file doesn't have a pdb format file")
            return None
        for line in self.list:
            if "STRAIN:" in line:
                res = line[line.find("STRAIN:")+8:]
                return(res[:res.find(";")])
        warnings.warn("This pdb file doesn't have this strain field.")
    
    def get_gene(self):
        """get gene of this pdb"""
        
        if self.list == None:
            warnings.warn("This pdb id file doesn't have a pdb format file")
            return None
        for line in self.list:
            if "GENE:" in line:
                res = line[line.find("GENE:")+6:]
                return(res[:res.find(";")])
        warnings.warn("This pdb file doesn't have this gene field.")
            
    def get_resolution(self):
        """get resolution of this pdb"""
        
        if self.list == None:
            warnings.warn("This pdb id file doesn't have a pdb format file")
            return None
        for line in self.list:
            if "RESOLUTION." in line and "ANGSTROMS." in line:
                num = line.find("RESOLUTION.") + 11
                while line[num] == " ":
                    num = num + 1
                    if line[num] != " ":
                        res = line[num:]
                        break
                return(res[:res.find("ANGSTROMS.")+9])
        warnings.warn("This pdb file doesn't have this resolution field.")
            
    def get_seqres(self):
        """get the seqres sequence of a pdb file"""
        
        if self.list == None:
            warnings.warn("This pdb id file doesn't have a pdb format file")
            return None
        list_se = []
        for line in self.list:
            if line[:6] == "SEQRES":
                list_se.append(line)
        if list_se == []:
            warnings.warn("This pdb file doesn't have Seqres sequence field.")
            return(None)
        list_se = [s.split() for s in list_se]
        df = pd.DataFrame(list_se)
        return(df)
    
    def get_atom_and_hetatm(self):
        """get the atom and hetatm sequence of a pdb file as a pandas dataframe"""
        
        if self.list == None:
            warnings.warn("This pdb id file doesn't have a pdb format file")
            return None
        list_atom = []
        # we only select line with atom and hetatm
        for line in self.list:
            if line[:4] == "ATOM" or line[:6] == "HETATM":
                list_atom.append(line)
        
        if list_atom == []:
            warnings.warn("This pdb file doesn't have atom and hetatm field.")
            return None
        list_s_atom = []
        # according to the string position to make a list of each row
        for line in list_atom:
            l = []
            if line[:6].isspace():
                l.append(None)
            else:
                l.append(line[:6].split()[0])
            if line[6:11].isspace():
                l.append(None)
            else:
                l.append(line[6:11].split()[0])
            
            if line[12:16].isspace():
                l.append(None)
            else:
                l.append(line[12:16].split()[0])
                
            if line[16].isspace():
                l.append(None)
            else:
                l.append(line[16].split()[0])
            
            if line[17:20].isspace():
                l.append(None)
            else:
                l.append(line[17:20].split()[0])
                
            if line[21].isspace():
                l.append(None)
            else:
                l.append(line[21].split()[0])
                
            if line[22:26].isspace():
                l.append(None)
            else:
                l.append(line[22:26].split()[0])
                
            if line[26].isspace():
                l.append(None)
            else:
                l.append(line[26].split()[0])
                
            if line[30:38].isspace():
                l.append(None)
            else:
                l.append(line[30:38].split()[0])
                
            if line[38:46].isspace():
                l.append(None)
            else:
                l.append(line[38:46].split()[0])
                
            if line[46:54].isspace():
                l.append(None)
            else:
                l.append(line[46:54].split()[0])
                
            if line[54:60].isspace():
                l.append(None)
            else:
                l.append(line[54:60].split()[0])
                
            if line[60:66].isspace():
                l.append(None)
            else:
                l.append(line[60:66].split()[0])
                
            if line[76:78].isspace():
                l.append(None)
            else:
                l.append(line[76:78].split()[0])
                
            if line[78:80].isspace():
                l.append(None)
            else:
                l.append(line[78:80].split()[0])
            
            list_s_atom.append(l)
                
        df = pd.DataFrame(list_s_atom)
        # use DataFrame function to convert a list to dataframe
        df.columns = ["Record Name", "serial", "name", "altLoc", "resName", "chainID", "resSeq", 
                     "iCode", "x", "y", "z", "occupancy", "tempFactor", "element", "charge"]
        return(df)
    
    def get_missing_residue(self):
        """get the missing residue as a pandas data frame"""
        
        if self.list == None:
            warnings.warn("This pdb id file doesn't have a pdb format file")
            return None
        
        res = None
        
        for line in self.list:
            if "MISSING RESIDUES" in line:
                res = line
        
        if res == None:
            warnings.warn("This pdb file doesn't have Missing Residue field.")
            return None
        
        inx = self.list.index(res)
        
        while True:
            if ("M RES C SSSEQI" not in self.list[inx]) and ("RES C SSSEQI" not in self.list[inx]):
                inx = inx + 1
            else:
                break
        
        if "M RES C SSSEQI" in self.list[inx]:
            flag = 1
        elif "RES C SSSEQI" in self.list[inx]:
            flag = 0
            
        inx += 1
        
        list_miss = []
        if flag:
        # according to the string position to make a list of each row
            while not self.list[inx][10:].isspace():
                l = []
                line = self.list[inx]
                if line[13].isspace():
                    l.append(None)
                else:
                    l.append(line[13])
                if line[15:18].isspace():
                    l.append(None)
                else:
                    l.append(line[15:18])
                if line[19].isspace():
                    l.append(None)
                else:
                    l.append(line[19])
                if line[21:26].isspace():
                    l.append(None)
                else:
                    l.append(line[21:26])
                if line[26].isspace():
                    l.append(None)
                else:
                    l.append(line[26])
                list_miss.append(l)
                inx = inx + 1
        else:
            while not self.list[inx][10:].isspace():
                l = []
                line = self.list[inx]

                if line[15:18].isspace():
                    l.append(None)
                else:
                    l.append(line[15:18])
                if line[19].isspace():
                    l.append(None)
                else:
                    l.append(line[19])
                if line[21:26].isspace():
                    l.append(None)
                else:
                    l.append(line[21:26])
                if line[26].isspace():
                    l.append(None)
                else:
                    l.append(line[26])
                list_miss.append(l)
                inx = inx + 1
        
        df = pd.DataFrame(list_miss)
        if flag:
            df.columns = ["MODEL NUMBER", "RESIDUE NAME", "CHAIN IDENTIFIER", "SEQUENCE NUMBER", "INSERTION CODE"]
        else:
            df.columns = ["RESIDUE NAME", "CHAIN IDENTIFIER", "SEQUENCE NUMBER", "INSERTION CODE"]
        return(df)
```

```python
def get_general_table(self):
    """get a general table of this pdb file"""
    
    if self.list == None:
        warnings.warn("This pdb id file doesn't have a pdb format file")
        return(pd.DataFrame({"PDB_ID": [self.id],
                       "Enzyme_Type": [None],
                       "Name": [None],
                       "Organism_Name": [None],
                       "Organism_Taxid": [None],
                       "Ec_Number": [None],
                       "Strain": [None],
                       "Gene": [None],
                       "Resolution": [None],
                       "Chain_Id": [None],
                       "Num_Seqres": [None],
                       "Seqres": [None],
                       "Num_Missing": [None],
                       "Missing_Residue": [None],
                       "Missing_Id": [None]
                       }))
    
    chain_id = self.get_chain_id().split(", ")
    num_chain = len(chain_id)
    
    PDB_ID = [self.id] * num_chain
    Enzyme_Type = [self.get_enzyme_type()] * num_chain
    Name = [self.get_name()] * num_chain
    Organism_Name = [self.get_organism_name()] * num_chain
    Organism_Taxid = [self.get_organism_taxid()] * num_chain
    Ec_Number = [self.get_ec_number()] * num_chain
    Strain = [self.get_strain()] * num_chain
    Gene = [self.get_gene()] * num_chain
    Resolution = [self.get_resolution()] * num_chain
    Seqres = self.get_seqres()
    
    Num_Seqres = []
    Seqres_total = []
    for iid in chain_id:
        se = Seqres[Seqres[2] == iid].loc[:, 4:].values.tolist()
        a = []
        [a.extend(i) for i in se]
        Num_Seqres.append(len(a))
        Seqres_total.append(a)
    
    Num_Missing = []
    Missing = self.get_missing_residue()
    Missing_Residue = []
    Missing_Id = []
    
    if isinstance(Missing, str) or Missing is None:
        Num_Missing = [None] * num_chain
        Missing_Residue = [None] * num_chain
        Missing_Id = [None] * num_chain
    else:
        for iid in chain_id:
            se = Missing[Missing["CHAIN IDENTIFIER"] == iid]
            a = se["RESIDUE NAME"].values.tolist()
            b = se["SEQUENCE NUMBER"].values.tolist()
            Num_Missing.append(len(a))
            Missing_Residue.append(a)
            Missing_Id.append(b)
    
    df = pd.DataFrame({"PDB_ID": PDB_ID,
                       "Enzyme_Type": Enzyme_Type,
                       "Name": Name,
                       "Organism_Name": Organism_Name,
                       "Organism_Taxid": Organism_Taxid,
                       "Ec_Number": Ec_Number,
                       "Strain": Strain,
                       "Gene": Gene,
                       "Resolution": Resolution,
                       "Chain_Id": chain_id,
                       "Num_Seqres": Num_Seqres,
                       "Seqres": Seqres_total,
                       "Num_Missing": Num_Missing,
                       "Missing_Residue": Missing_Residue,
                       "Missing_Id": Missing_Id
                       })
    return(df)
```

```python
def get_entity_id(self):
    """get the entity id and belonging chains of each pdb file"""
    
    L = self.list
    dic = {}
    for i in range(len(L)):
        if re.search("^COMPND.*MOL_ID:", L[i]):
            j = i + 1
            while not re.search("^COMPND.*CHAIN:", L[j]):
                j += 1
            dic[L[i][L[i].find("MOL_ID: ")+8:L[i].find(";")]] = L[j][L[j].find("CHAIN: ")+7:L[j].find(";")]
    return dic

pdb.get_entity_id = get_entity_id
```

```python
def get_part_mod_resi(self):
    """get the partially modeled residue index of fasta sequence of each chain in one pdb file"""
    
    pre = "{annotations(\
                   reference:PDB_INSTANCE,\
                   sources:PDB_INSTANCE,\
    queryId:\""
    suf = "\"\
  ){\
    target_id\
    features{\
      feature_id\
      type\
      provenance_source\
      description\
      name\
      feature_positions{\
        beg_seq_id\
        end_seq_id\
      }\
    }\
  }\
}"
    chain = [x for x in self.get_chain_id().split(", ")]
    pre_url = "https://1d-coordinates.rcsb.org/graphql?query="
    dic = {}
    for iid in chain:
        full_url = pre_url + urllib.parse.quote(pre + self.id.upper() + "." + iid + suf)
        text = requests.get(full_url).text
        start = text.find("partially modeled residue") + 50
        end = text.find("feature_id", start) - 3
        yellow = text[start:end]
        yellow_num = re.findall('\d+', yellow)
        yellow_num = [int(x) for x in yellow_num]
        yellow_num
        num = []
        for i in range(0, len(yellow_num), 2):
            num.append((yellow_num[i], yellow_num[i+1]))
        res = []
        for i,j in num:
            if i == j:
                res.append(i)
            elif i < j:
                res.extend([x for x in range(i, j+1)])
        dic[iid] = res
    return dic
pdb.get_part_mod_resi = get_part_mod_resi
```

```python
def get_unmod_resi(self):
    """get the unmodeled residue index of fasta sequence of each chain in one pdb file"""
    
    pre = "{annotations(\
                   reference:PDB_INSTANCE,\
                   sources:PDB_INSTANCE,\
    queryId:\""
    suf = "\"\
  ){\
    target_id\
    features{\
      feature_id\
      type\
      provenance_source\
      description\
      name\
      feature_positions{\
        beg_seq_id\
        end_seq_id\
      }\
    }\
  }\
}"
    chain = [x for x in self.get_chain_id().split(", ")]
    pre_url = "https://1d-coordinates.rcsb.org/graphql?query="
    dic = {}
    for iid in chain:
        full_url = pre_url + urllib.parse.quote(pre + self.id.upper() + "." + iid + suf)
        text = requests.get(full_url).text
        start = text.find("unmodeled residue") + 40
        end = text.find("feature_id", start) - 3
        yellow = text[start:end]
        yellow_num = re.findall('\d+', yellow)
        yellow_num = [int(x) for x in yellow_num]
        yellow_num
        num = []
        for i in range(0, len(yellow_num), 2):
            num.append((yellow_num[i], yellow_num[i+1]))
        res = []
        for i,j in num:
            if i == j:
                res.append(i)
            elif i < j:
                res.extend([x for x in range(i, j+1)])
        dic[iid] = res
    return dic
pdb.get_unmod_resi = get_unmod_resi
```

## Add this yellow and grey box index into the `general_table()` as two separate columns

```python
def get_general_table(self):
    """get a general table of this pdb file"""
    
    if self.list == None:
        warnings.warn("This pdb id file doesn't have a pdb format file")
        return(pd.DataFrame({"PDB_ID": [self.id],
                       "Enzyme_Type": [None],
                       "Name": [None],
                       "Mutation": [None],
                       "Global Stoichiometry": [None],
                       "Organism_Name": [None],
                       "Organism_Taxid": [None],
                       "Ec_Number": [None],
                       "Strain": [None],
                       "Gene": [None],
                       "Resolution": [None],
                       "Chain_Id": [None],
                       "Fasta Sequence": [None],
                       "Partial Modeled Residue(yellow)": [None],
                       "Unmodeled Residue(grey)": [None],
                       "Entity ID": [None],
                       "Num_Seqres": [None],
                       "Seqres": [None],
                       "Num_Missing": [None],
                       "Missing_Residue": [None],
                       "Missing_Id": [None]
                       }))
    
    chain_id = self.get_chain_id().split(", ")
    num_chain = len(chain_id)
    
    PDB_ID = [self.id.upper()] * num_chain
    Enzyme_Type = [self.get_enzyme_type()] * num_chain
    Name = [self.get_name()] * num_chain
    Organism_Name = [self.get_organism_name()] * num_chain
    Organism_Taxid = [self.get_organism_taxid()] * num_chain
    Ec_Number = [self.get_ec_number()] * num_chain
    Strain = [self.get_strain()] * num_chain
    Gene = [self.get_gene()] * num_chain
    Resolution = [self.get_resolution()] * num_chain
    Seqres = self.get_seqres()
    
    Num_Seqres = []
    Seqres_total = []
    for iid in chain_id:
        se = Seqres[Seqres[2] == iid].loc[:, 4:].values.tolist()
        a = []
        [a.extend(i) for i in se]
        Num_Seqres.append(self.get_num_chain()[iid])
        Seqres_total.append(a)
    
    Num_Missing = []
    Missing = self.get_missing_residue()
    Missing_Residue = []
    Missing_Id = []
    
    if isinstance(Missing, str) or Missing is None:
        Num_Missing = [None] * num_chain
        Missing_Residue = [None] * num_chain
        Missing_Id = [None] * num_chain
    else:
        for iid in chain_id:
            se = Missing[Missing["CHAIN IDENTIFIER"] == iid]
            a = se["RESIDUE NAME"].values.tolist()
            b = se["SEQUENCE NUMBER"].values.tolist()
            Num_Missing.append(len(a))
            Missing_Residue.append(a)
            Missing_Id.append(b)
    
    fasta = []
    dic = self.get_fasta()
    for iid in chain_id:
        fasta.extend([v for k,v in dic.items() if iid in k.lstrip("Chains ")])
    
    entity_id = []
    dic_1 = self.get_entity_id()
    for iid in chain_id:
        entity_id.extend([k for k, v in dic_1.items() if iid in v])
    
    partial_index = []
    dic_2 = self.get_part_mod_resi()
    for iid in chain_id:
        partial_index.extend([v for k, v in dic_2.items() if iid in k])
    
    unmodel = []
    dic_3 = self.get_unmod_resi()
    for iid in chain_id:
        unmodel.extend([v for k, v in dic_3.items() if iid in k])
        
    df = pd.DataFrame({"PDB_ID": PDB_ID,
                       "Enzyme_Type": Enzyme_Type,
                       "Name": Name,
                       "Mutation": [self.get_mutation()] * num_chain,
                       "Global Stoichiometry": [self.get_glo_sto()] * num_chain,
                       "Organism_Name": Organism_Name,
                       "Organism_Taxid": Organism_Taxid,
                       "Ec_Number": Ec_Number,
                       "Strain": Strain,
                       "Gene": Gene,
                       "Resolution": Resolution,
                       "Chain_Id": chain_id,
                       "Fasta Sequence": fasta,
                       "Partial Modeled Residue(yellow)": partial_index,
                       "Unmodeled Residue(grey)": unmodel,
                       "Entity ID": entity_id,
                       "Num_Seqres": Num_Seqres,
                       "Seqres": Seqres_total,
                       "Num_Missing": Num_Missing,
                       "Missing_Residue": Missing_Residue,
                       "Missing_Id": Missing_Id
                       })
    return(df)

pdb.get_general_table = get_general_table
```
